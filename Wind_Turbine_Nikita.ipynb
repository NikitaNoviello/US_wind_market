## Wind data

```python
#reset index
wind_data.set_index('case_id', inplace=True)
```

```python
#changed dtype
wind_data['t_img_date'] = pd.to_datetime(wind_data.t_img_date, dayfirst=False)
```

```python
#dropped columns
wind_clean = wind_data.drop(columns=['faa_ors', 'faa_asn', 'usgs_pr_id', 't_fips'])
```

```python
#rename columns
wind_clean.rename(columns={'t_state':'state'
,'t_county':'county'
,'p_name':'wind_farm'
,'p_year':'year'
,'p_tnum':'turbine_count'
,'p_cap':'project_capacity'
,'t_manu':'manufact'
,'t_model':'model'
,'t_cap':'turbine_capacity'
,'t_hh':'hub_height'
,'t_rd':'rotor_diameter'
,'t_rsa':'rotor_swept_area'
,'t_ttlh':'total_height'
,'retrofit':'retrofit'
,'retrofit_year':'retrofit_year'
,'t_conf_atr':'conf_atr'
,'t_conf_loc':'conf_loc'
,'t_img_date':'img_date'
,'t_img_srce':'img_source'
,'xlong':'long'
,'ylat':'lat'},inplace=True)
```

```python
#drop year nulls 
wind_clean.dropna(subset=['year'], inplace=True)
```
